# Task 2

```python
# Extracting first 2 seconds of the signal

import soundfile as sf
data, sr = sf.read(filename)
duration = 2
num_samples = int(duration * sr)
first_two_seconds = data[:num_samples]
output_filename = 'first_2sec.wav'
sf.write(output_filename, first_two_seconds, sr)
```

```python
# Normalizing the extracted values

first_2_sec_normalized = librosa.util.normalize(first_two_seconds)

plt.figure(figsize = (14,7))
librosa.display.waveshow(first_2_sec_normalized, sr = sample_rate)
plt.title("Waveform of first 2 sec")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

# Amplification

```python
amplification_value = 2
first_2_sec_amplified = first_2_sec_normalized * amplification_value


plt.figure(figsize = (14,7))
librosa.display.waveshow(first_2_sec_amplified, sr = sample_rate)
plt.title("Waveform of Amplified first 2 sec")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

# Deamplification


```python
de_amplification_value = 0.5
first_2_sec_deamplified = first_2_sec_normalized * de_amplification_value

plt.figure(figsize=(14, 7))
librosa.display.waveshow(first_2_sec_deamplified, sr=sample_rate)
plt.title("Waveform of De-amplified First 2 Seconds")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

# UpSampling

```python
up_sample_rate = sample_rate * 2
first_2_sec_up_sampled = librosa.resample(first_2_sec_normalized, orig_sr = sample_rate, target_sr = up_sample_rate)

plt.figure(figsize=(14, 7))
librosa.display.waveshow(first_2_sec_up_sampled, sr=up_sample_rate)
plt.title("Waveform of Up-sampled First 2 Seconds")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

# DownSampling

```python
down_sample_rate = sample_rate // 2
first_2_sec_down_sampled = librosa.resample(first_2_sec_normalized, orig_sr = sample_rate, target_sr = down_sample_rate)

plt.figure(figsize=(14, 7))
librosa.display.waveshow(first_2_sec_down_sampled, sr=down_sample_rate)
plt.title("Waveform of Down-sampled First 2 Seconds")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

```
